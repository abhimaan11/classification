# train and test the data and split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
Sc=StandardScaler()
```

```python
x_train=Sc.fit_transform(x_train)
```

```python
x_test=Sc.transform(x_test)
```

# model selection 

```python
from sklearn.model_selection import cross_val_score #for testing crossvalscore it will show the diff between ypridicted and ytest
```

```python
from sklearn.model_selection import StratifiedKFold #stratified means different parts of data run(test will be shuffled)
```

```python
models=[]
```

```python
models.append(('LR',LogisticRegression(solver='liblinear',multi_class='ovr')))   #ovr one vs rest
```

```python
models.append(('LDA',LinearDiscriminantAnalysis()))  
```

```python
models.append(('KNN',KNeighborsClassifier()))
```

```python
models.append(('DT',DecisionTreeClassifier()))
```

```python
models.append(('NB',GaussianNB()))
```

```python
models.append(('svm',SVC(gamma='auto')))   
```

```python
results=[]
names=[]
res=[]
```

```python
for name,model in models:
    KFold=StratifiedKFold(n_splits=10,random_state=None)
    CV_results=cross_val_score(model,x_train,y_train,cv=KFold,scoring='accuracy')
    names.append(name)
    res.append(CV_results.mean())
    print(name,CV_results.mean())
```

```python
import matplotlib.pyplot as pyplot
```

```python
results = []
names = []
res = []
for name, model in models:
    kfold = StratifiedKFold(n_splits=10, random_state=None)
    cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    res.append(cv_results.mean())
    print('%s: %f' % (name, cv_results.mean()))

pyplot.ylim(.900, .999)
pyplot.bar(names, res, color ='blue', width = 0.6)

pyplot.title('Algorithm Comparison')
pyplot.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
